# Random Forest Regression

## Importing the libraries

```
# =========================================
# Importing the Libraries
# =========================================

# NumPy for numerical operations
import numpy as np

# Matplotlib for visualization
import matplotlib.pyplot as plt

# Pandas for handling datasets
import pandas as pd
```

## Importing the dataset

```
# =========================================
# Importing the Dataset
# =========================================

# Read the CSV file containing Position, Level and Salary
dataset = pd.read_csv('Position_Salaries.csv')

# ================================
# Prepare Feature Matrix (X)
# ================================
# iloc[:, 1:-1] → take all rows, columns from index 1 to the second-last column
# In our dataset:
#   - Column 0: Position (categorical, not used as numeric feature)
#   - Column 1: Level (used as numeric feature)
X = dataset.iloc[:, 1:-1].values

# ================================
# Prepare Dependent Variable Vector (y)
# ================================
# iloc[:, -1] → take all rows, last column
# Column 2: Salary (target value to predict)
y = dataset.iloc[:, -1].values
```

## Training the Random Forest Regression model on the whole dataset

```
# =========================================
# Training the Random Forest Regression Model
# =========================================

# Import Random Forest Regressor from scikit-learn
from sklearn.ensemble import RandomForestRegressor

# Create the regressor
# n_estimators → number of trees in the forest
# random_state → ensures reproducibility of results
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)

# Fit the regressor to the dataset
# X → feature matrix (Level)
# y → target vector (Salary)
regressor.fit(X, y)
```

## Predicting a new result

```
# =========================================
# Predicting a new result with Random Forest
# =========================================

# Predict the Salary for a specific Level (6.5)
# regressor → trained Random Forest Regressor
# Input must be 2D array: [[level]]
prediction = regressor.predict([[6.5]])

# Display the predicted Salary
print(prediction)
```

## Visualising the Random Forest Regression results (higher resolution)

```
# ================================================
# Visualising the Random Forest Regression Results (High Resolution)
# ================================================

# Create a high-resolution grid for X (step = 0.01)
# This allows us to see the "smooth" prediction curve clearly
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))

# Plot original data points
plt.scatter(X, y, color = 'red', label='Actual data')

# Plot Random Forest predictions on high-resolution grid
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue', label='Random Forest Prediction')

# Add title and axis labels
plt.title('Truth or Bluff (Random Forest Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
